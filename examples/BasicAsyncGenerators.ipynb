```python
import asyncio

async def inf(s, start=0):
    i = start
    while True:
        await asyncio.sleep(s)
        yield i
        i += 1
```

```python
c1 = inf(0.1)
```

```python
c2 = inf(1.0)
```

```python
c3 = c2 + 1
```
